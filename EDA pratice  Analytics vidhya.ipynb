```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.pandas.set_option('display.max_columns',None)
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
#create  sub plots
fig,axes = plt.subplots(2,2,figsize=(16,9))

#Plot1: Gender of Student
sns.countplot(data=df,x='gender',ax=axes[0,0])
axes[0,0].set_title('Gender of student')
axes[0,0].set_xticklabels(axes[0,0].get_xticklabels(),rotation=0)

# Plot 2: Race/ethnicity of students
sns.countplot(data=df,x='race/ethnicity',ax=axes[0,1])
axes[0,1].set_title('Race/ethnicity of students')
axes[0,1].set_xticklabels(axes[0,1].get_xticklabels(),rotation=0)

# Plot 3: Lunch status of students
sns.countplot(data=df, x='lunch', ax=axes[1, 0])
axes[1, 0].set_title('Lunch status of students')
axes[1, 0].set_xticklabels(axes[1, 0].get_xticklabels(), rotation=0)

# Plot 4: Test preparation course
sns.countplot(data=df, x='test preparation course', ax=axes[1, 1])
axes[1, 1].set_title('Test preparation course')
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=0)

#adjusting spacing between subplots
plt.tight_layout()

#showplot
plt.show()


```

```python
#we will be making a boxplot of the numerical columns (math score, reading score, and writing score) in the dataset.
df.boxplot()
```

```python
#We will now make a distribution plot of the math score of the students. 
sns.displot(df['math score'],kde=True)
```

```python
# We will now look at the correlation between the 3 scores with the help of a heatmap.
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,square=True)
plt.yticks(rotation=0)
plt.show()
```

```python
#We will now move on to Bivariate Analysis. We will look at a relational plot in Seaborn.
sns.relplot(x='math score',y='writing score',hue='gender',data=df)
```
