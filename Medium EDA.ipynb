```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load Data
df=pd.read_csv("nyc_weather.csv")
pd.pandas.set_option('display.max_columns',None)
```

```python
sns.distplot(df.Humidity.dropna(),kde=False, bins=39)
```

```python
df
sns.set(style="whitegrid")
ax = sns.boxplot(x=df.Temperature,showfliers=False)
```
